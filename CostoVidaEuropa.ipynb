```python
import pandas as pd
df = pd.read_excel("/content/cheapest_countries_to_study_Europe_2023.xlsx")
df.info()
df.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,16))
plt.subplot(3,1,1)
plt.plot(df["Country"],df["Total yearly living costs and fees (£)"],c="skyblue")
plt.title("Costo total anual")
plt.xticks(df["Country"],rotation =90)
plt.subplot(3,1,2)
plt.plot(df["Country"],df["Average yearly tuition fees (£)"],c="red")
plt.title("Costos de estudios anual")
plt.xticks(df["Country"],rotation =90)
plt.subplot(3,1,3)
plt.plot(df["Country"],df["Yearly student living costs (£)"],c="green")
plt.title("Costos de vida anual")
plt.xticks(df["Country"],rotation =90)
plt.subplots_adjust(hspace=0.5)
plt.show()
```
