```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
df = pd.read_csv("/content/predictive_maintenance.csv")
df = df.set_index("UDI")
df.info()
df.head()
```

```python
for i in df:
  print(df[i].value_counts())
```

```python
filtro = df["Failure Type"]=="No Failure"
df1 = df[~filtro]
df1
```

De la grafica de violin y de caja podemos apreciar que las fallas por desgaste de la herramienta (Tool Wear Failure) y las por sobreesfuerzo (Overstrain Failure) son las que se presentan a más de 175 horas de uso, significando que la vida util de las herramientas es en promedio 205 horas.

```python
px.violin(df1, y="Tool wear [min]", color="Failure Type", title="Distribución de las horas de funcionamiento por tipo de falla",box=True)
```

Hipótesis: la temperatura de las fallas y no fallas no hay diferencia significativa en las medias

```python
from scipy import stats
df2 = df[filtro]
```

Podemos verificar que el p_value es menor que el 5% por lo que se RECHAZA la hipótess de igualdad de media de temeperaturas, o sea si hay diferencia significativa en la media de la temperatura

```python
t_statistic, p_value = stats.ttest_ind(df1["Process temperature [K]"], df2["Process temperature [K]"])
print(t_statistic, p_value)
```

Hipótesis: el torque de las fallas y no fallas no hay diferencia significativa en las medias.
Se RECHAZA la hipotesis debido a que p-value es menor al 5%

```python
t_statistic, p_value = stats.ttest_ind(df1["Torque [Nm]"], df2["Torque [Nm]"])
print(t_statistic, p_value)
```

Con las siguientes figuras corroboramos que las medias de las fallas y no fallas de la temperatura y del torque no son las misma, indicando un factor importante que las fallas pueden ser originadas por altas temperaturas o torques.

```python
df3 = df.replace("Heat Dissipation Failure","Failure")
df3 = df3.replace("Power Failure","Failure")
df3 = df3.replace("Tool Wear Failure","Failure")
df3 = df3.replace("Overstrain Failure","Failure")
df3 = df3.replace("Random Failures","Failure")
figura1 = px.box(df3, y="Process temperature [K]", color="Failure Type", title="Distribución de la temperatura del proceso por falla o no falla")
figura1.show()
figura2 = px.box(df3, y="Torque [Nm]", color="Failure Type", title="Distribución del torque del proceso por falla o no falla")
figura2.show()
```
